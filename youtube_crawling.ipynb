```python
#자동 스크롤링
pages = 200
elem1= dr.find_element(By.TAG_NAME,'body')
while pages:
    elem1.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.5)
    pages -= 1
```

```python
#검색된 동영상 수 확인
contents=dr.find_elements(By.CSS_SELECTOR,'#contents > ytd-video-renderer')
len(contents)
```

```python
#썸네일,제목,조회수,기간,유튜버,장르 저장해둘 리스트 생성
thumbnails=[]
titles=[]
views=[]
timelines=[]
youtubers=[]
genre=[]
```

```python
#썸네일,제목,조회수,기간,유튜버,장르 저장
for content in contents:
        thumbnail=content.find_element(By.ID,'thumbnail').find_element(By.TAG_NAME,'img').get_attribute('src')
        thumbnails.append(thumbnail)
        title=content.find_element(By.ID,'video-title').text
        titles.append(title)
        view=content.find_element(By.ID,'metadata-line').text.split('\n')[0].split(' ')[1]
        views.append(view)
        timeline=content.find_element(By.ID,'metadata-line').text.split('\n')[1]
        timelines.append(timeline)
        youtuber=content.find_element(By.ID,'channel-info').find_element(By.CSS_SELECTOR,'#text > a').text
        youtubers.append(youtuber)
        genre.append(keyword)
```

```python
#조회수 문자->숫자로 변경
views_int=[]
for view in views:
    view=view[:-1]
    if type(view[-1])==str:
        if view[-1]=='억':
            view=view[:-1]
            view=int(float(view)*100000000)
        elif view[-1]=='천':
            view=view[:-1]
            view=int(float(view)*1000)
        elif view[-1]=='만':
            view=view[:-1]
            view=int(float(view)*10000)
        else:
            view=int(view)
    views_int.append(view)        
views_int
```

```python
#데이터프레임으로 변환
df=pd.DataFrame((zip(titles,views_int,timelines,youtubers,genre,thumbnails)),columns=['title','view','timeline','youtuber','genre','thumbnail'])
```

```python
#조회수 10000 이상 영상만 추출
df1=df[df['view']>=10000]
```

```python
df1.reset_index(inplace=True,drop=True)
df1
```
