### 패션

#### OKT: 동사, 형용사, 부사, 알파벳

```python
fashion=pd.read_csv('/content/drive/MyDrive/DScover_A조/메인 프로젝트/주제 크롤링/패션.csv')
```

```python
okt = Okt()
sentences_tag = []
for sentence in fashion['title']:
  morph = okt.pos(sentence)
  sentences_tag.append(morph)
  print(morph)
```

```python
adj_list = []
verb_list = []
adverb_list = []
alpha_list = []

for my_sentence in sentences_tag:
  for word, tag in my_sentence:
    if tag in ['Adjective']:
      adj_list.append(word)
    if tag in ['Verb']:
      verb_list.append(word)
    if tag in ['Adverb']:
      adverb_list.append(word)
    if tag in ['Alpha']:
      alpha_list.append(word)  
```

```python
print(adj_list)
print(verb_list)
print(adverb_list)
print(alpha_list)
```

```python
from collections import Counter
adj_counts = Counter(adj_list)
adj_counts.most_common(20)
```

```python
verb_counts = Counter(verb_list)
verb_counts.most_common(20)
```

```python
adverb_counts = Counter(adverb_list)
adverb_counts.most_common(10)
```

```python
alpha_counts = Counter(alpha_list)
alpha_counts.most_common(20)
```

#### KKMA: 명사

```python
from konlpy.tag import Kkma
kkma = Kkma()
noun_list=[]
for i in fashion['title']:
  noun_list.append(kkma.nouns(i))
print(noun_list)
```

```python
dic={}
for i in noun_list:
  for word in i:
    if word not in dic:
      dic[word]=1
    else:
      dic[word]+=1
print(dic)
```

```python
kkma_d1 = sorted(dic.items(), key=lambda x: x[1], reverse=True)
kkma_d2=[item for item in kkma_d1 if item[1]>=10]
print(kkma_d2)
```

```python
stopwords=['패션', '대', '하', '2', '김', 'ㅣ', '나', '년', '4', '10', '수', '1', '5', '리', '총', '2022년']
df=[item for item in kkma_d2 if item[0] not in stopwords]
df
```

# 유사도 분석

### 패션

```python
fashion=pd.read_csv('/content/drive/MyDrive/DScover_A조/메인 프로젝트/주제 크롤링/패션.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(min_df =2 , ngram_range=(1,2))
# min_df-몇개의 문서에서 등장했는지, 
# ngram 1=단어 1개씩 묶음(unigram) 
#       2=단어 2개를 한 묶음으로 취급(bigram)
```

```python
features=vectorizer.fit_transform(fashion['title'])
features.todense()
```

```python
feature_names=vectorizer.get_feature_names_out()
```

```python
import numpy as np
dtm_np = np.array(features.todense()) # dtm document term metrix
```

```python
pd.DataFrame(data=dtm_np, columns=feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim=cosine_similarity(dtm_np,dtm_np)
```

```python
cosine_sim.shape
```

```python
text='코디'
title_in=[]
for i,title in enumerate(fashion['title']):
    if text in title:
        title_in.append((i,title))
title_in  
```

```python
#idx = 조회수가 높은 타이틀 순서
idx=title_in[0][0]
sim_scores=list(enumerate(cosine_sim[idx]))
```

```python
sim_scores=sorted(sim_scores,key=lambda x :x[1], reverse=True)
```

```python
sim_scores=sim_scores[0:11]
sim_scores
```

```python
title_indices=[i[0] for i in sim_scores]
title_indices
```

```python
fashion.iloc[title_indices]['title']
```

```python
import pandas as pd
import numpy as np
from string import punctuation

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
fashion = pd.read_csv('/content/drive/MyDrive/DScover_A조/메인 프로젝트/주제 크롤링/패션.csv')
```

```python
print('열의 개수: ',len(fashion.columns))
print(fashion.columns)
```

```python
print(fashion['title'].isnull().values.any())
```

```python
title = []
# 헤드라인의 값들을 리스트로 저장
title.extend(list(fashion.title.values)) 
title[:5]
```

```python
print('총 샘플의 개수 : {}'.format(len(title))) # 현재 샘플의 개수
```

```python
 # Unknown 값을 가진 샘플 제거
title = [word for word in title if word != "Unknown"]
print('노이즈값 제거 후 샘플의 개수 : {}'.format(len(title)))
```

```python
def repreprocessing(raw_sentence):
    preproceseed_sentence = raw_sentence.encode("utf8").decode("ascii",'ignore')
    # 구두점 제거와 동시에 소문자화
    return ''.join(word for word in preproceseed_sentence if word not in punctuation).lower()

preporcessed_headline = [repreprocessing(x) for x in title]
preporcessed_headline[:5]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(title)
vocab_size = len(tokenizer.word_index) + 1
print('단어 집합의 크기 : %d' % vocab_size)
```

```python
sequences = list()

for sentence in title:

    # 각 샘플에 대한 정수 인코딩
    encoded = tokenizer.texts_to_sequences([sentence])[0] 
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)

sequences[:11]
```

```python
index_to_word = {}
for key, value in tokenizer.word_index.items(): # 인덱스를 단어로 바꾸기 위해 index_to_word를 생성
    index_to_word[value] = key

print('빈도수 상위 582번 단어 : {}'.format(index_to_word[582]))
```

```python
max_len = max(len(l) for l in sequences)
print('샘플의 최대 길이 : {}'.format(max_len))
```

```python
sequences = pad_sequences(sequences, maxlen=max_len, padding='pre')
print(sequences[:3])
```

```python
sequences = np.array(sequences)
X = sequences[:,:-1]
y = sequences[:,-1]
```

```python
print(X[:3])
```

```python
print(y[:3])
```

```python
y = to_categorical(y, num_classes=vocab_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM
```

```python
embedding_dim = 10
hidden_units = 128

model = Sequential()
model.add(Embedding(vocab_size, embedding_dim))
model.add(LSTM(hidden_units))
model.add(Dense(vocab_size, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=200, verbose=2)
```

```python
def sentence_generation(model, tokenizer, current_word, n): # 모델, 토크나이저, 현재 단어, 반복할 횟수
    init_word = current_word
    sentence = ''

    # n번 반복
    for _ in range(n):
        encoded = tokenizer.texts_to_sequences([current_word])[0]
        encoded = pad_sequences([encoded], maxlen=max_len-1, padding='pre')

        # 입력한 X(현재 단어)에 대해서 y를 예측하고 y(예측한 단어)를 result에 저장.
        result = model.predict(encoded, verbose=0)
        result = np.argmax(result, axis=1)

        for word, index in tokenizer.word_index.items(): 
            # 만약 예측한 단어와 인덱스와 동일한 단어가 있다면
            if index == result:
                break

        # 현재 단어 + ' ' + 예측 단어를 현재 단어로 변경
        current_word = current_word + ' '  + word

        # 예측 단어를 문장에 저장
        sentence = sentence + ' ' + word

    sentence = init_word + sentence
    return sentence
```

```python
def repreprocessing(raw_sentence):
    preproceseed_sentence = raw_sentence.encode("utf8").decode("ascii",'ignore')
    # 구두점 제거와 동시에 소문자화
    return ''.join(word for word in preproceseed_sentence if word not in punctuation).lower()

preporcessed_headline = [repreprocessing(x) for x in title]
preporcessed_headline[:5]
```

```python
sequences = list()

for sentence in title:

    # 각 샘플에 대한 정수 인코딩
    encoded = tokenizer.texts_to_sequences([sentence])[0] 
    for i in range(1, len(encoded)):
        sequence = encoded[:i+1]
        sequences.append(sequence)

sequences[:11]
```

```python
index_to_word = {}
for key, value in tokenizer.word_index.items(): # 인덱스를 단어로 바꾸기 위해 index_to_word를 생성
    index_to_word[value] = key

print('빈도수 상위 582번 단어 : {}'.format(index_to_word[582]))
```
